```python
import pandas as pd
import matplotlib.pyplot as plt 

# dados = pd.read_csv('Artigo WIT.csv')
dados = pd.read_csv('TotalMeninas.csv', delimiter=';')
dadosTotais = pd.read_csv('Totalinscritos.csv')
```

```python
def salvaDuplicadas(df):
    return df[df.duplicated()].values.tolist()
```

```python
dados2022 = dados['Nomes-2022']
dados2023 = dados['Nomes-2023']
dados2024 = dados['Nomes-2024']

# dados2022 = dados['Nº Inscrição-2022']
# dados2023 = dados['Nº Inscrição-2023']
# dados2024 = dados['Nº Inscrição-2024']


dados2022 = dados2022.dropna()
dados2023 = dados2023.dropna()
dados2024 = dados2024.dropna()

print("> Nº de inscritos mulheres em 2022: ", dados2022.count())
print("> Nº de inscritos mulheres em 2023: ", dados2023.count())
print("> Nº de inscritos mulheres em 2024: ", dados2024.count())

##Salvando o dados duplicados para usar depois:
duplicatas_total = []
for i in [dados2022, dados2023, dados2024]:
    duplicatas_total.append(salvaDuplicadas(i))

print('----------------------------')
print('DUPLICADAS: \n', duplicatas_total)
print('----------------------------')

dados2022 = dados2022.drop_duplicates()
dados2023 = dados2023.drop_duplicates()
dados2024 = dados2024.drop_duplicates()

numGirls2022 = dados2022.count()
numGirls2023 = dados2023.count()
numGirls2024 = dados2024.count()

print("Nº de inscritos mulheres em 2022: ", numGirls2022)
print("Nº de inscritos mulheres em 2023: ", numGirls2023)
print("Nº de inscritos mulheres em 2024: ", numGirls2024)
```

```python
anos = [2022, 2023, 2024]
num_meninas = [numGirls2022, numGirls2023, numGirls2024]

cores = ['#92b7c0', '#2b3955', '#d94448']

# plt.figure(figsize=(10, 6))
# plt.bar(anos, num_meninas, color=cores)
# plt.title('Meninas')
# plt.ylabel('Nº de inscritas')
# plt.xlabel('Anos')
# plt.xticks(anos)
# plt.show()

plt.figure(figsize=(10, 6))

# Criando as barras
bars = plt.bar(anos, num_meninas, color=cores)

# Adicionando rótulos no topo de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, round(yval, 1), ha='center', va='bottom')

# plt.title('Meninas')
plt.ylabel('Nº de inscritas')
plt.xticks(anos)
plt.xlabel('Anos')
plt.show()
```

```python
dadosTotais2022 = dadosTotais['Nº Inscrição-2022']
dadosTotais2023 = dadosTotais['Nº Inscrição-2023']
dadosTotais2024 = dadosTotais['Nº Inscrição-2024']

dadosTotais2022 = dadosTotais2022.dropna()
dadosTotais2023 = dadosTotais2023.dropna()
dadosTotais2024 = dadosTotais2024.dropna()

print("> Total inscritos em 2022: ", dadosTotais2022.count())
print("> Total inscritos em 2023: ", dadosTotais2023.count())
print("> Total inscritos em 2024: ", dadosTotais2024.count())

dadosTotais2022 = dadosTotais2022.drop_duplicates()
dadosTotais2023 = dadosTotais2023.drop_duplicates()
dadosTotais2024 = dadosTotais2024.drop_duplicates()

dadosTotais2022 = dadosTotais2022.count()
dadosTotais2023 = dadosTotais2023.count()
dadosTotais2024 = dadosTotais2024.count()

print("Total de inscritos em 2022: ", dadosTotais2022)
print("Total de inscritos em 2023: ", dadosTotais2023)
print("Total de inscritos em 2024: ", dadosTotais2024)
```

```python
NumHomens2022= dadosTotais2022 - numGirls2022
NumHomens2023= dadosTotais2023 - numGirls2023
NumHomens2024= dadosTotais2024 - numGirls2024

print("Nº de inscritos homens em 2022: ", NumHomens2022)
print("Nº de inscritos homens em 2023: ", NumHomens2023)
print("Nº de inscritos homens em 2024: ", NumHomens2024)
```
